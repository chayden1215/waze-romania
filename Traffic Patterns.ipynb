```python
import dask.dataframe as dd
import geopandas as gpd
import h3
import pandas as pd
import pydeck as pdk
from shapely import wkt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
```

```python
data_path = r'C:\Users\noah2\OneDrive\Desktop\GSE 580\Final Model\Github Corrections\waze-romania\data'
df = pd.read_csv(data_path + "/fullclean.csv")
```

```python
plot = df.groupby(['year', 'month'],as_index=False)['uuid'].count()
#plot = plot[plot.year!=2019]
sns.set_theme(style="whitegrid")
ax = sns.barplot(x='month', y='uuid', hue='year', data=plot)
```

```python
plot = df.groupby(['type', 'month'],as_index=False)['uuid'].count()
sns.set_theme(style="whitegrid")
ax = sns.barplot(x='month', y='uuid', hue='type', data=plot)
```

```python
plot = df[df.year !=2019].groupby(['rush', 'month'],as_index=False)['uuid'].count()
sns.set_theme(style="whitegrid")
ax = sns.barplot(x='month', y='uuid', hue='rush', data=plot)
```

```python
plot = df.groupby(['year', 'month', 'rush'],as_index=False)['uuid'].count()
g = sns.FacetGrid(plot, col="rush")
g.map_dataframe(sns.barplot, x="month", y="uuid")
```

```python
g.add_legend()
plot = df.groupby(['hour'],as_index=False)['uuid'].count()
sns.set_theme(style="whitegrid")
ax = sns.barplot(x='hour', y='uuid', data=plot)
```

```python
#too small!
plot = df.groupby(['hour', 'month'],as_index=False)['uuid'].count()
g = sns.FacetGrid(plot, col="month", height=10)
g.map_dataframe(sns.barplot, x="hour", y="uuid")
```
